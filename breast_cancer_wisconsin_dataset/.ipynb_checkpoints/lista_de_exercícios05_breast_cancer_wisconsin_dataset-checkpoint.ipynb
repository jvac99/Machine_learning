#### Preparando o ambiente

**Importação das bibliotecas**

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
from sklearn import model_selection
from sklearn import linear_model
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn import ensemble
from sklearn import tree
from sklearn import svm
from sklearn import neighbors
from sklearn import naive_bayes
from sklearn.model_selection import train_test_split
```

**Carregando e definindo as colunas do dataset**

```python
data = pd.read_csv('breast_cancer_wisconsin_dataset.csv').drop(columns=['Unnamed: 32'])
```

#### Trabalhando com os dados

**Criação do dataset de validação**

```python
# Features
X = data.loc[:,['id', 'diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]
# Labels
y = data.loc[:,['diagnosis']]
```

```python
dataset = data.values
# Features
X = dataset[:,2:]
# Labels
y = dataset[:,1]
```

* Treino (60%) e Teste (40%) - randomizado, mas não estratificado
* Treino (60%) e Teste (40%) - randomizado e estratificado
* Treino (80%) e Teste (20%) - randomizado, mas não estratificado
* Treino (80%) e Teste (20%) - randomizado e estratificado

```python
train_test_list = []
train_test_list.append(train_test_split(X, y, train_size=0.6, random_state=42)) 
train_test_list.append(train_test_split(X, y, train_size=0.6, random_state=42, stratify=y))
train_test_list.append(train_test_split(X, y, train_size=0.8, random_state=42)) 
train_test_list.append(train_test_split(X, y, train_size=0.8, random_state=42, stratify=y))
```

#### Criação dos modelos

Detecção de câncer de mama - Conjunto de dados disponível em: https://www.kaggle.com/datasets/uciml/breast-cancer-wisconsin-data 
Em caso de dúvida, veja também o Capítulo 2 do livro Mãos à obra: aprendizado de máquina com Scikit-Learn, Keras & TensorFlow.

**Classificadores utilizados:**


* DecisionTreeClassifier

* SVC

* KNN

* Logistic Regression

* Naive Bayes

* SGDClassifier

* RandomForest

* GradientBoosting

```python
modelos = {}
```

```python
modelos['DT'] = tree.DecisionTreeClassifier()
modelos['SVM'] = svm.SVC()
modelos['KNN'] = neighbors.KNeighborsClassifier()
modelos['LogReg'] = linear_model.LogisticRegression(multi_class='multinomial',solver='newton-cg')
modelos['NB'] = naive_bayes.GaussianNB()
modelos['SGD'] = linear_model.SGDClassifier()
modelos['RF'] = ensemble.RandomForestClassifier()
modelos['GB'] = ensemble.GradientBoostingClassifier()
```

#### Resultados

* Compare os resultados de acurácia de diferentes algoritmos de classificação para cada um dos cenários a seguir:

```python
results = {
    'result_1_tr60_r_n_e': None,
    'result_2_tr60_r_e': None,
    'result_3_tr80_r_n_e': None,
    'result_4_tr80_r_e': None,
}
```

```python
# Criando um DataFrame
dict_resultados = {'algs': list(modelos.keys())} 
#
my_zip = zip(train_test_list, results.keys())

#
for (X_train, X_test, y_train, y_test), rkey in my_zip:
    resultados_train = []
    resultados_test = []
    for modelo_str in modelos.keys():
        modelo = modelos[modelo_str]
        modelo.fit(X_train, y_train)
        y_pred_train = modelo.predict(X_train)
        y_pred_test = modelo.predict(X_test)
        acuracia_train = metrics.accuracy_score(y_train, y_pred_train)
        acuracia_test = metrics.accuracy_score(y_test, y_pred_test)
        resultados_train.append(round(acuracia_train, 2))
        resultados_test.append(round(acuracia_test, 2))
    dict_resultados.update({f'acc_train_{rkey}': resultados_train, f'acc_test_{rkey}': resultados_test})
```

* Compare os resultados de acurácia de diferentes algoritmos de classificação para cada um dos cenários a seguir:

```python
df_resultados = pd.DataFrame(dict_resultados)
```
