1. Implemente uma função Python para cada uma das seguintes métricas de regressão:

* MAE

* MSE

* RMSE (Raiz quadrada do MSE)

Cada função deve receber valores reais (y) e valores preditos (y_pred) como parâmetros de entrada e retornar o valor calculado da métrica. Esse parâmetros podem ser arrays do numpy ou DataFrames do Pandas. O ideal é realizar as operações de forma vetorizada, evitando acessar cada célula individualmente. Não usar as implementações já existentes do Scikit-Learn.

```python
import numpy as np
import pandas as pd
import cmath
import math
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import train_test_split
```

```python

def MAE (y_verdadeiros, y_estimados):
  y_dif = np.array(y_verdadeiros) - np.array(y_estimados)
  return sum(abs(y_dif))/len(y_verdadeiros)
```

```python
# Exemplo
mat_a = [3, -0.5, 2, 7]
mat_b = [2.5, 0.0, 2, 8]
print(MAE(mat_a, mat_b))
print(mean_absolute_error(mat_a, mat_b))
```

```python
def MSE(y_verdadeiros, y_estimados):
  y_dif = np.array(y_verdadeiros) - np.array(y_estimados)
  return sum(y_dif ** 2)/len(y_verdadeiros)
```

```python
# Exemplo
print(MSE(mat_a, mat_b))
print(mean_squared_error(mat_a, mat_b, squared=True))
```

```python
# Raiz quadrada
def RMSE (mat_a, mat_b):
  return MSE(mat_a, mat_b) ** 0.5
```

```python
# Exemplo
print(RMSE(mat_a, mat_b))
print(mean_squared_error(mat_a, mat_b, squared=False))
```

2. Implemente uma função do Python para dividir de forma randomizada o dataset entre treino e teste. A função recebe como parâmetros:

* X - matriz com features

* y - vetor com rótulos

* perc_treino - percentual dos dados usados para treino seed - semente de randomização

A função deve retornar: X_train, y_train, X_test, y_test.  Não usar as implementações já existentes do Scikit-Learn.

```python
def traino_e_teste (X, y, perc_treino = 0.75):
  n_train = int(round(len(y) * perc_treino))
  X_train = X[:n_train]
  y_train = y[:n_train]
  X_test = X[n_train:]
  y_test = y[n_train:]
  return [X_train, y_train, X_test, y_test]
```

```python
data = np.random.rand(10, 3)
df = pd.DataFrame(data, columns=['x1', 'x2', 'y'])
df
```

```python
X = df [df.columns [:-1]]
X
```

```python
y = df [df.columns [-1:]]
y
```

```python
train_test_split(X, y)
```

```python
traino_e_teste(X, y)[3]
```
