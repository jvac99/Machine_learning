```python
data = pd.read_csv("heart.csv")
data.head()
```

```python
X = data.drop('target', axis=1).values
#scaler = StandardScaler()
#X1 = scaler.fit_transform(X)
X
```

```python
y = data['target']
y.head()
```

```python
# Dividindo em conjunto de train e test
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

2. Treine e avalie (de acordo com a métrica F1-Score), usando NB e SVM implementados por você, o problema de classificação binária a seguir: Heart Disease Dataset

```python
def my_models (random_state=42):  
  models = {}
  models['SVCS'] = SVC(random_state=random_state)
  models['SVM'] = svm.SVM()
  models['NBM'] = MultinomialNB()
  models['NBS'] = naivebayes.NaiveBayes()
  models['GNBS'] = GaussianNB()
  models['LR'] = logistic_regression.LogisticRegression()
  models['LRS'] = LogisticRegression()
  return models
```

```python
results_train_f1_micro = []
results_test_f1_micro = []
results_train_f1_macro = []
results_test_f1_macro = []
results_train_f1_weighted = []
results_test_f1_weighted = []
algs = []

models = my_models()


for models_str in models.keys():
    model = models[models_str]

    # Treinando o model
    model.fit(X_train, y_train)
    model.fit(X_test, y_test)

    # Predizendo
    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)

    # Avaliando o modelo
    algs.append(models_str)

    f1_score_micro_train = f1_score(y_test, y_pred_test, average='micro')
    f1_score_micro_test = f1_score(y_train, y_pred_train, average='micro')

    f1_score_macro_train = f1_score(y_test, y_pred_test, average='macro')
    f1_score_macro_test = f1_score(y_train, y_pred_train, average='macro')

    f1_score_weighted_train =  f1_score(y_test, y_pred_test, average='weighted')
    f1_score_weighted_test =  f1_score(y_train, y_pred_train, average='weighted')
    
    results_train_f1_micro.append(round(f1_score_micro_train, 2))
    results_test_f1_micro.append(round(f1_score_micro_test, 2))

    results_train_f1_macro.append(round(f1_score_macro_train, 2))
    results_test_f1_macro.append(round(f1_score_macro_test, 2))

    results_train_f1_weighted.append(round(f1_score_weighted_train, 2))
    results_test_f1_weighted.append(round(f1_score_weighted_test, 2))
    
df_result = pd.DataFrame({
    'algs': algs,
    'f1_train_micro': results_train_f1_micro,
    'f1_train_weighted': results_train_f1_weighted,
    'f1_train_macro': results_train_f1_macro,
    'f1_test_micro': results_test_f1_micro, 
    'f1_test_macro': results_test_f1_macro, 
    'f1_test_weighted': results_test_f1_weighted
})

df_result
```

```python
def my_models (random_state=42):  
  models = {}
  models['SVCS'] = SVC(random_state=random_state)
  models['SVM'] = svm.SVM()
  models['NBM'] = MultinomialNB()
  models['NBS'] = naivebayes.NaiveBayes()
  models['GNBS'] = GaussianNB()
  models['LR'] = logistic_regression.LogisticRegression()
  models['LRS'] = LogisticRegression()
  return models
```

```python
models = my_models()

param1 = {}
param1['classifier__C'] = [10**-2, 10**-1, 10**0, 10**1, 10**2]
param1['classifier__kernel'] = ['linear', 'poly', 'rbf', 'sigmoid', 'precomputed']
param1['classifier__class_weight'] = [None, {0:1,1:5}, {0:1,1:10}, {0:1,1:25}]
param1['classifier'] = [models['SVCS']]

param2 = {}
param2['classifier'] = [models['SVM']]

param3 = {}
param3['classifier__alpha'] = [10**0, 10**1, 10**2]
param3['classifier'] = [models['NBM']]

param4 = {}
param4['classifier'] = [models['NBS']]

param5 = {}
param5['classifier'] = [models['GNBS']]

param6 = {}
param6['classifier'] = [models['LR']]

param7 = {}
param7['classifier__C'] = [10**-2, 10**-1, 10**0, 10**1, 10**2]
param7['classifier__penalty'] = ['l1', 'l2']
param7['classifier'] = [models['LRS']]



params = [param1, param2, param3, param4, param5, param6, param7]
```

```python
gs_results_train_f1_micro = []
gs_results_test_f1_micro = []
gs_results_train_f1_macro = []
gs_results_test_f1_macro = []
results_train_f1_weighted = []
gs_results_test_f1_weighted = []
gs_best_params = []
gs_best_score = []


algs = []

for models_str in models.keys():
    model = models[models_str]

    pipeline = Pipeline([('standard', StandardScaler()), ('classifier', model)])

    gs = GridSearchCV(pipeline, params, cv=3, n_jobs=-1, scoring='f1_micro', return_train_score=True)

    # Treinando
    gs.fit(X_train, y_train)

    gs_best_params.append({'algs': models_str, 'best_score': gs.best_params_})

    gs_best_score.append({'algs': models_str, 'best_score': gs.best_score_})

    y_pred_gs = gs.predict(X_test)

    #print(classification_report(y_test, y_pred_gs, digits=4))

      # Avaliando o modelo
    algs.append(models_str)

    f1_score_micro_train = f1_score(y_test, y_pred_test, average='micro')
    f1_score_micro_test = f1_score(y_train, y_pred_train, average='micro')

    f1_score_macro_train = f1_score(y_test, y_pred_test, average='macro')
    f1_score_macro_test = f1_score(y_train, y_pred_train, average='macro')

    f1_score_weighted_train =  f1_score(y_test, y_pred_test, average='weighted')
    f1_score_weighted_test =  f1_score(y_train, y_pred_train, average='weighted')
    
    gs_results_train_f1_micro.append(round(f1_score_micro_train, 2))
    gs_results_test_f1_micro.append(round(f1_score_micro_test, 2))

    gs_results_train_f1_macro.append(round(f1_score_macro_train, 2))
    gs_results_test_f1_macro.append(round(f1_score_macro_test, 2))

    results_train_f1_weighted.append(round(f1_score_weighted_train, 2))
    gs_results_test_f1_weighted.append(round(f1_score_weighted_test, 2))
    
gs_df_result = pd.DataFrame({
    'algs': algs,
    'gs_f1_train_micro': gs_results_train_f1_micro,
    'gs_f1_train_weighted': results_train_f1_weighted,
    'gs_f1_train_macro': gs_results_train_f1_macro,
    'gs_f1_test_micro': gs_results_test_f1_micro, 
    'gs_f1_test_macro': gs_results_test_f1_macro, 
    'gs_f1_test_weighted': gs_results_test_f1_weighted
})

gs_df_result
```
