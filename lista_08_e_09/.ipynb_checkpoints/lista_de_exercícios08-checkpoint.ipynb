3. Treine e avalie, usando as métricas e algoritmos implementados por você, um problema simples de regressão com apenas uma variável sobre o conjunto de dados Boston House Prices disponível no próprio scikit learn (datasets.load_boston()). Use como feature somente a feature RM, que corresponde ao número médio de quartos por habitação. O rótulo (label ou target) é o atributo medv, que é o valor médio das casas ocupadas pelos proprietários em $1000s. Use 80% do dataset para treino e o restante para teste.

```python
data = load_boston()
```

```python
data.keys()
```

```python
print(data['DESCR'])
```

```python
df = pd.DataFrame(data.data)
df
```

```python
df.info()
```

```python
df.columns = data['feature_names']
df
```

```python
df['class'] = data.target
df
```

```python
df['class'].value_counts()
```

```python
(df['class'].value_counts() / df.shape[0] * 100).round(2)
```

```python
X = df.iloc[:, :-1]
#X = df['RM']
X.head()
```

```python
y = df['class']
y.head()
```

```python
# Dividindo em conjunto de train e test
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
```

```python
model = KNeighborsRegressor(n_neighbors=3)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(metrics.mae(y_pred, y_test))
print(mean_absolute_error(y_pred, y_test))
print(metrics.mse(y_pred, y_test))
print(mean_squared_error(y_pred, y_test))
print(metrics.rmse(y_pred, y_test))
print(mean_squared_error(y_pred, y_test, squared=False))
```

```python
model = knnre.KNNRE(3)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(metrics.mae(y_pred, y_test))
print(mean_absolute_error(y_pred, y_test))
print(metrics.mse(y_pred, y_test))
print(mean_squared_error(y_pred, y_test))
print(metrics.rmse(y_pred, y_test))
print(mean_squared_error(y_pred, y_test, squared=False))
```

```python
regressor = linear_regression.LinearRegression(learning_rate=0.01, n_iters=100)
regressor.fit(X_train, y_train)
predictions = regressor.predict(X_test)

#mse = metrics.mse(y_test, predictions)
#print("MSE:", mse)
```

```python
regressor = regressao_linear_simples.RegressaoLinearSimples(x=X_train, y=X_test)
regressor.fit(X_train, y_train)
predictions = regressor.predict(X_test)

#mse = mse(y_test, predictions)
#print("MSE:", mse)
```

```python
k = 5
clf = knnr.KNNR(k=k)
clf.fit(X_train, y_train)
clf.predict(X_test)
```

```python
from sklearn import datasets

def accuracy(y_true, y_pred):
    accuracy = np.sum(y_true == y_pred) / len(y_true)
    return accuracy

iris = datasets.load_iris()
X, y = iris.data, iris.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1234
)
```

```python
model = KNeighborsRegressor(n_neighbors=3)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(metrics.mae(y_pred, y_test))
print(mean_absolute_error(y_pred, y_test))
print(metrics.mse(y_pred, y_test))
print(mean_squared_error(y_pred, y_test))
print(metrics.rmse(y_pred, y_test))
print(mean_squared_error(y_pred, y_test, squared=False))
```

```python
k = 3
model = knnc.KNNC(k=k)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(metrics.mae(y_pred, y_test))
print(mean_absolute_error(y_pred, y_test))
print(metrics.mse(y_pred, y_test))
print(mean_squared_error(y_pred, y_test))
print(metrics.rmse(y_pred, y_test))
print(mean_squared_error(y_pred, y_test, squared=False))
```

```python
k = 3
model = knnca.KNNCA(k=k)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(metrics.mae(y_pred, y_test))
print(mean_absolute_error(y_pred, y_test))
print(metrics.mse(y_pred, y_test))
print(mean_squared_error(y_pred, y_test))
print(metrics.rmse(y_pred, y_test))
print(mean_squared_error(y_pred, y_test, squared=False))
```
