```python
#y = df.loc[:,'class']
y = boston.target
```

```python
# Dividindo em conjunto de train e test
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
def my_models (random_state=42, k = 3):  
  models = {}
  models['KNNS'] = KNeighborsRegressor(n_neighbors=k)
  models['KNNR'] = knnr.KNNR(k=k)
  models['KNNRE'] = knnre.KNNRE(k=k)
  models['LRS'] = LinearRegression()
  return models
```

```python
results_train_mae = []
results_test_mae = []
results_train_mse = []
results_test_mse = []
results_train_rmse = []
results_test_rmse = []
algs = []

models = my_models()


for models_str in models.keys():
    model = models[models_str]

    # Treinando o model
    model.fit(X_train, y_train)
    model.fit(X_test, y_test)

    # Predizendo
    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)

    # Avaliando o modelo
    mae_train = metrics.mae(y_train, y_pred_train)
    mae_test = metrics.mae(y_test, y_pred_test)

    mae_sklearn_train = mean_absolute_error(y_train, y_pred_train)
    mae_sklearn_test = mean_absolute_error(y_test, y_pred_test)

    print('Mae: ', mae_train, ' ', mae_sklearn_train)
    print('Mae sklearn: ', mae_test, ' ', mae_sklearn_test)

    mse_train = metrics.mse(y_train, y_pred_train)
    mse_test = metrics.mse(y_test, y_pred_test)

    mse_sklearn_train = mean_squared_error(y_train, y_pred_train)
    mse_sklearn_test = mean_squared_error(y_test, y_pred_test)

    print('Mse: ', mse_train, ' ', mse_sklearn_train)
    print('Mse sklearn: ', mse_test, ' ', mse_sklearn_test)

    rmse_train = metrics.rmse(y_train, y_pred_train)
    rmse_test = metrics.rmse(y_test, y_pred_test)

    rmse_sklearn_train = mean_squared_error(y_train, y_pred_train, squared=False)
    rmse_sklearn_test = mean_squared_error(y_test, y_pred_test, squared=False)

    print('Rmse: ', rmse_train, ' ', rmse_sklearn_train)
    print('Rme sklearn: ', rmse_test, ' ', rmse_sklearn_test)

    algs.append(models_str)

    results_train_mae.append(round(mae_train, 2))
    results_test_mae.append(round(mae_test, 2))
    results_train_mse.append(round(mse_train, 2))
    results_test_mse.append(round(mse_test, 2))
    results_train_rmse.append(round(rmse_train, 2))
    results_test_rmse.append(round(rmse_test, 2))

df_result = pd.DataFrame({
    'algs': algs,
    'mae_train': results_train_mae,
    'mae_test': results_test_mae,
    'mse_train': results_train_mse,
    'mse_test': results_test_mse,
    'rmse_train': results_train_rmse,
    'rmse_test': results_test_rmse
})

df_result
```

    * SGDRegressor (teste alguns valores dos seguintes hiper-parâmetros: penalty, alpha, max_iter, early_stopping)

```python
X_standard = StandardScaler().fit_transform(X)
```

```python
# Dividindo em conjunto de train e test
X_train, X_test, y_train, y_test = train_test_split(X_standard, y, train_size=0.8, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
clf1 = LinearRegression()
clf2 = SGDRegressor()
clf3 = Ridge()
clf4 = Lasso()
clf5 = ElasticNet()
```

```python
param1 = {}


param2 = {}
param2['classifier__penalty'] = ['l1', 'l2', 'elasticnet', 'none']
param2['classifier__alfa'] = [0, 1]
param2['classifier__max_iter'] = [10, 20, 40, 80, 100]
param2['classifier__early_stopping'] = [True, False]


param3 = {}
param3['classifier__alfa'] = [0, 1]
param3['classifier__max_iter'] = [10, 20, 40, 80, 100]


param4 = {}
param4['classifier__alfa'] = [0, 1]
param4['classifier__max_iter'] = [10, 20, 40, 80, 100]


param5 = {}
param5['classifier__alfa'] = [0, 1]
param5['classifier__max_iter'] = [10, 20, 40, 80, 100]
param5['classifier__l1_ratio'] = [10**-2, 10**-1, 10**0, 10**1, 10**2]

params = [param1, param2, param3, param4, param5]
```

```python
pipeline = Pipeline([('scaler', StandardScaler()), ('rg', clf1)])

gs = GridSearchCV(pipeline, [param3], cv=3, n_jobs=-1, scoring='f1', return_train_score=True)

# Treinando
gs.fit(X_train, y_train)

print(gs.best_params_)

print(gs.best_score_)

y_pred_gs = gs.predict(X_test)

print(classification_report(y_test, y_pred_gs, digits=4))

print("F1 Score Micro:",f1_score(y_pred_gs, y_test, average='micro'))
print("F1 Score Macro:",f1_score(y_pred_gs, y_test, average='macro'))
print("F1 Score Weighted:",f1_score(y_pred_gs, y_test, average='weighted'))
```
